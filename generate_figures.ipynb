```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# LFW
k_vals = [1, 2, 3, 5, 7, 9, 12, 15, 25, 50]
err = [0.8524, 0.8736, 0.8769, 0.8732, 0.8672, 0.8615, 0.8582, 0.8588, 0.8619, 0.8717]
err_convolved = [0.8345, 0.8625, 0.8631, 0.8588, 0.8551, 0.8512, 0.8528, 0.8499, 0.8526, 0.8662]

plt.plot(k_vals, err, label = "kNN", marker='o', linestyle='--', color='cornflowerblue')
plt.plot(k_vals, err_convolved, label = "Convolution + kNN", marker='o', linestyle='--', color='orange')
#plt.plot(1, 0.8434, label = "kNN (k=1)", marker='*', color='red', linestyle='')
#plt.plot(1, 0.8181, label = "Convolution + kNN (k=1)", marker='*', color='green', linestyle='')
plt.xlabel("k")
plt.ylabel("CV Error")
plt.title("k-Fold Cross Validation for\n\"Labeled Faces in the Wild\"") #5-fold
plt.legend()
plt.savefig("CV_LFW.pdf")
plt.show()
```
